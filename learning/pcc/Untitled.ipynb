```python
sentence_list = ['I am king Robert of the house Baratheon, first of my name',
                 'I am king Joffrey of the house Baratheon, first of my name and plus my father in truth is actually my uncle',
                 
                 'I am King Aegon Moondragon, fifth of my name.']
lower_sentence_list = [x.lower() for x in sentence_list]
print(lower_sentence_list)
```

```python
lower_sentence_list = []
for x in sentence_list:
    x.lower()
    lower_sentence_list.append(x.lower())
print(lower_sentence_list)
```
